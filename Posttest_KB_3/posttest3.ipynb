# Import Library

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

# Membaca file dataset

```python
df = pd.read_csv('fifa_eda_stats.csv')
df.head(10)
```

```python
df.shape
```

# Diagram Pie Potential

```python
data_pie = df['Potential'].value_counts()
```

```python
df.info()
```

```python
data_pie = df['Potential'].value_counts().reset_index(name = "potential")
```

```python
plt.pie(data_pie.potential,
    labels = data_pie["potential"], 
    startangle=110, 
    autopct='%.1f%%', 
    shadow=True
)
plt.title('Diagram Pie Potential Player')
plt.show()
```

# Korelasi Atribut Dengan Tipe Angka

```python
plt.figure(figsize=(25,20))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap='flare'
) 
plt.title('korelasi antar attribut variabel ')

plt.show()
```

```python
plt.figure(figsize=(15,10))

x = df['Overall']
y = df['Potential']

plt.bar(x, y, color='green')

plt.title('Perbandingan Vehicle Year dan Renter Trips Taken', size=16)
plt.ylabel('Renter Trips Taken', size=14)
plt.xlabel('Vehicle Year', size=14)
plt.xticks(size=12)
plt.yticks(size=12)

plt.show()
```
